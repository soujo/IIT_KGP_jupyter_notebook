# 1.0 Search in Arrays

```python
import numpy as np
import time
import matplotlib.pyplot as plt

# ..........................................Function for binary search........................................

def bin_search(arr, key):
    p0 = time.perf_counter()  # Record initial time
    arr.sort()  # Sort the array in ascending order
    l = 0  # Initialize lower bound index
    u = len(arr) - 1  # Initialize upper bound index
    found = False  # Flag to track if the key is found
    while l <= u and not found:
        mid = (l + u) // 2  # Calculate middle index
        if key == arr[mid]:  # If the key is found at middle index
            found = True
        elif key > arr[mid]:  # If key is greater, search the right half
            l = mid + 1
        else:  # If key is smaller, search the left half
            u = mid - 1
    p1 = time.perf_counter()  # Record final time
    return p1 - p0  # Return time taken for binary search

# List of array sizes for analysis
arr_sizes = [20, 30, 40, 50, 60, 70, 80, 90, 100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]
avg_time_bin = []  # List to store average times for binary search
length_bin = []    # List to store corresponding array lengths

# Loop through different array sizes
for size in arr_sizes:
    A = np.random.randint(1, 1000, size)  # Generate random array of specified size
    length_bin.append(size)  # Store current array size
    time_taken = np.zeros(size)  # Initialize an array to store times for each search
    for i in range(size):
        key = A[i]  # Get the key for the current iteration
        time_taken[i] = bin_search(A, key)  # Perform binary search and record time
    avg_time_bin.append(sum(time_taken) / size)  # Calculate and store average time for this array size


# ..........................................Function for Linear search........................................    
    
    
def lin_search(arr, key):
    t0 = time.perf_counter()  # Record initial time
    for i in range(len(arr)):
        if arr[i] == key:  # If the key is found at index i
            break  # Exit the loop since the key is found
    t1 = time.perf_counter()  # Record final time
    return t1 - t0  # Return time taken for linear search

avg_time_lin = []  # List to store average times for linear search
length_lin = []    # List to store corresponding array lengths

# Loop through different array sizes
for size in arr_sizes:
    A = np.random.randint(1, 1000, size)  # Generate random array of specified size
    length_lin.append(size)  # Store current array size
    time_taken = np.zeros(size)  # Initialize an array to store times for each search
    for i in range(size):
        key = A[i]  # Get the key for the current iteration
        time_taken[i] = lin_search(A, key)  # Perform linear search and record time
    avg_time_lin.append(sum(time_taken) / size)  # Calculate and store average time for this array size

# Create a plot to visualize the results
fig, ax = plt.subplots()
ax.plot(length_bin, avg_time_bin, 'mo-', label='binary search')  # Plot binary search results
ax.plot(length_lin, avg_time_lin, 'co-', label='linear search')   # Plot linear search results
ax.set_xlabel('Length (Number of Elements) of the Arrays')
ax.set_ylabel('Average Time taken by Searching Algorithms (Seconds)')
ax.set_title('Average Time Vs Array Size Performance for Searching Algorithms')
plt.legend()  # Display legend to differentiate between plots
     
```

# 2.0 Sorting in Arrays 

```python
import numpy as np
import time
import matplotlib.pyplot as plt

# ..........................................Function for Linear Sort........................................

def linear_sort(a):
    n0 = time.perf_counter()
    for i in range(1,len(a)):       
        key = i
        while a[key-1] > a[key] and key > 0:   # here this case says that if the left element of the pointer is greater than the element pointed by the pointer,
            a[key], a[key-1] = a[key - 1], a[key]   # then the numbers are swaped. 
            key = key -1                       # it will moving the pointer to left till pointer = 0
    n1 = time.perf_counter()
    return n1-n0                               #here we have used this case in such a way that n1-n0 is time taken for sorting                                    

tavg_linear = np.zeros(19)   #it will initialize the array to store avg. time complexity
length_linear = np.zeros(19) #it will initialize the array to store corresponding length of array

for i in range(1,10):        
    x = np.random.randint(1,1000,i*10)   # here we are taking arrays of length 10, 20,....100
    t0 = np.zeros(10)
    for j in range(1,10):
        t0[j] = linear_sort(x)           # we have considered 10 instances of worst time complexity for each array
    tavg_linear[i] = sum(t0)/10          #here it will be there for average time complexity
    length_linear[i] = len(x) 
for i in range(2,10):
    x = np.random.randint(1,1000,i*100)  #here we have taken arrays of length 200,300,....1000
    t1 = np.zeros(10)
    for j in range(1,10):
        t1[j] = linear_sort(x)           # here we have considered 10 instances of worst time complexity for each array
    tavg_linear[i+9] = sum(t1)/10
    length_linear[i+9] = len(x)
    
    
#..................................... Function for Bubble Sort..........................................
    
    
# Function to perform bubble sort
def bubble_sort(a):
    n2 = time.perf_counter()  # Record initial time
    for i in range(len(a)):
        for j in range(0, len(a) - i - 1):  # Inner loop for subsequent iterations
            if a[j] > a[j + 1]:  # If an element is greater than its right neighbor
                a[j], a[j + 1] = a[j + 1], a[j]  # Swap the elements
    n3 = time.perf_counter()  # Record final time
    return n3 - n2  # Return time taken for sorting

tavg_bubble = np.zeros(19)   # Initialize an array to store average time complexities for bubble sort
length_bubble = np.zeros(19)  # Initialize an array to store corresponding array lengths

# Loop through different array sizes
for i in range(1, 10):
    y = np.random.randint(1, 1000, i * 10)  # Generate random array of specified size
    t0 = np.zeros(10)  # Initialize an array to store times for each sort
    for j in range(1, 10):
        t0[j] = bubble_sort(y)  # Perform bubble sort and record time
    tavg_bubble[i] = sum(t0) / 10  # Calculate and store average time for this array size
    length_bubble[i] = len(y)

# Repeat the loop for larger array sizes
for i in range(2, 10):
    y = np.random.randint(1, 1000, i * 100)  # Generate random array of specified size
    t1 = np.zeros(10)  # Initialize an array to store times for each sort
    for j in range(1, 10):
        t1[j] = bubble_sort(y)  # Perform bubble sort and record time
    tavg_bubble[i + 9] = sum(t1) / 10  # Calculate and store average time for this array size
    length_bubble[i + 9] = len(y)

# Create a plot to visualize the results
fig, ax = plt.subplots()
ax.plot(length_bubble, tavg_bubble, 'go-', label='bubble sort')  # Plot bubble sort results
ax.plot(length_linear, tavg_linear, 'ro-', label="linear_sort")  # Plot linear sort results
ax.set_xlabel('Length (Number of Elements) of the Arrays')
ax.set_ylabel('Average Time taken by sorting Algorithms (Seconds')
ax.set_title('Average Time Vs Array Size Performance for Sorting Algorithms')
plt.legend()  # Display legend to differentiate between plots


        
```
