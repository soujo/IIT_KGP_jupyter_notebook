```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
F1 = 50  # Fundamental frequency in Hz
Fs = 1500  # Sampling frequency in Hz
T = 1  # Duration in seconds

# Angular frequencies
Omega = 2 * np.pi * F1
Omega_3F1 = 3 * Omega

# Time vector
t = np.arange(0, T, 1/Fs)

# Signal s1(t)
s1_t = np.sin(Omega * t) + (1/3) * np.sin(Omega_3F1 * t)

# Number of samples
N = len(s1_t)

# Compute the DFT using the formula
S1_f = np.zeros(N, dtype=complex)
for k in range(N):
    for n in range(N):
        S1_f[k] += s1_t[n] * np.exp(-1j * 2 * np.pi * k * n / N)

# Frequency vector
f = np.fft.fftfreq(N, 1/Fs)  # Can still use fftfreq to get the frequency vector

# Magnitude spectrum
magnitude_spectrum = np.abs(S1_f)

# Phase spectrum
phase_spectrum = np.angle(S1_f)

# Plot the magnitude spectrum
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(f, magnitude_spectrum)
plt.title('Magnitude Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)

# Plot the phase spectrum
plt.subplot(2, 1, 2)
plt.plot(f, phase_spectrum)
plt.title('Phase Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Phase (Radians)')
plt.grid(True)

# Display the plots
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
F1 = 50  # Fundamental frequency in Hz
Fs = 1500  # Sampling frequency in Hz
T = 1  # Duration in seconds

# Angular frequencies
Omega = 2 * np.pi * F1
Omega_3F1 = 3 * Omega

# Time vector
t = np.arange(0, T, 1/Fs)

# Signal s1(t)
s1_t = np.sin(Omega * t) + (1/3) * np.sin(Omega_3F1 * t)

# Number of samples
N = len(s1_t)

# DCT-II implementation
def dct_ii(x):
    N = len(x)
    X = np.zeros(N)
    for k in range(N):
        for n in range(N):
            X[k] += x[n] * np.cos(np.pi / N * (n + 0.5) * k)
    return X

# DST-II implementation
def dst_ii(x):
    N = len(x)
    Y = np.zeros(N)
    for k in range(N):
        for n in range(N):
            Y[k] += x[n] * np.sin(np.pi / N * (n + 0.5) * (k + 1))
    return Y

# Compute DCT-II and DST-II
dct_coeffs = dct_ii(s1_t)
dst_coeffs = dst_ii(s1_t)

# Magnitude spectra
dct_magnitude_spectrum = np.abs(dct_coeffs)
dst_magnitude_spectrum = np.abs(dst_coeffs)

# Frequency vector
f = np.arange(N) * Fs / N

# Plot the magnitude spectrum of DCT-II
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(f, dct_magnitude_spectrum)
plt.title('Magnitude Spectrum of DCT-II')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)

# Plot the magnitude spectrum of DST-II
plt.subplot(2, 1, 2)
plt.plot(f, dst_magnitude_spectrum)
plt.title('Magnitude Spectrum of DST-II')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)

# Display the plots
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of n
N = 40  # Length of the signal
n = np.arange(N)

# Define the unit step function u[n]
u = np.heaviside(n, 1)

# Define the signal s2[n]
s2 = (0.5 ** n) * u

# Compute the DFT of s2[n]
def dft(x):
    N = len(x)
    X = np.zeros(N, dtype=complex)
    for k in range(N):
        for m in range(N):
            X[k] += x[m] * np.exp(-1j * 2 * np.pi * k * m / N)
    return X

# Compute the IDFT of s2[n]
def idft(X):
    N = len(X)
    x = np.zeros(N, dtype=complex)
    for n in range(N):
        for k in range(N):
            x[n] += X[k] * np.exp(1j * 2 * np.pi * k * n / N)
    return x / N

# Compute the DCT-II of s2[n]
def dct_ii(x):
    N = len(x)
    X = np.zeros(N)
    for k in range(N):
        for n in range(N):
            X[k] += x[n] * np.cos(np.pi / N * (n + 0.5) * k)
    return X

# Compute the IDCT-II of s2[n]
def idct_ii(X):
    N = len(X)
    x = np.zeros(N)
    for n in range(N):
        x[n] = 0.5 * X[0]
        for k in range(1, N):
            x[n] += X[k] * np.cos(np.pi / N * (n + 0.5) * k)
    return x

# Compute the DST-II of s2[n]
def dst_ii(x):
    N = len(x)
    Y = np.zeros(N)
    for k in range(N):
        for n in range(N):
            Y[k] += x[n] * np.sin(np.pi / N * (n + 0.5) * (k + 1))
    return Y

# Compute the IDST-II of s2[n]
def idst_ii(Y):
    N = len(Y)
    x = np.zeros(N)
    for n in range(N):
        for k in range(N):
            x[n] += Y[k] * np.sin(np.pi / N * (n + 0.5) * (k + 1))
    return x

# Function to retain a given percentage of coefficients
def retain_coefficients(coefficients, percentage):
    N = len(coefficients)
    retained = np.zeros(N, dtype=coefficients.dtype)
    threshold = int(np.ceil(percentage * N))
    retained[:threshold] = coefficients[:threshold]
    return retained

# Calculate the DFT, DCT-II, and DST-II of s2[n]
dft_s2 = dft(s2)
dct_s2 = dct_ii(s2)
dst_s2 = dst_ii(s2)

# Retain 90% of the coefficients
percentage = 0.90

# Retain 75% of the coefficients
# percentage = 0.75

# Retain 45% of the coefficients
# percentage = 0.45

dft_s2_retained = retain_coefficients(dft_s2, percentage)
dct_s2_retained = retain_coefficients(dct_s2, percentage)
dst_s2_retained = retain_coefficients(dst_s2, percentage)

# Reconstruct the signal using the inverse transforms
s2_dft_reconstructed = idft(dft_s2_retained).real
s2_dct_reconstructed = idct_ii(dct_s2_retained)
s2_dst_reconstructed = idst_ii(dst_s2_retained)

# Compute the Mean Squared Error (MSE)
mse_dft = np.mean((s2 - s2_dft_reconstructed) ** 2)
mse_dct = np.mean((s2 - s2_dct_reconstructed) ** 2)
mse_dst = np.mean((s2 - s2_dst_reconstructed) ** 2)

# Print the MSE values
print(f"MSE for DFT reconstruction: {mse_dft}")
print(f"MSE for DCT reconstruction: {mse_dct}")
print(f"MSE for DST reconstruction: {mse_dst}")

# Plot the original and reconstructed signals
plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.stem(n, s2, linefmt='b-', markerfmt='bo', basefmt=' ', label='Original signal')
plt.stem(n, s2_dft_reconstructed, linefmt='r--', markerfmt='ro', basefmt=' ', label='Reconstructed signal (DFT)')
plt.title('Original and Reconstructed Signal (DFT)')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)

plt.subplot(3, 1, 2)
plt.stem(n, s2, linefmt='b-', markerfmt='bo', basefmt=' ', label='Original signal')
plt.stem(n, s2_dct_reconstructed, linefmt='r--', markerfmt='ro', basefmt=' ', label='Reconstructed signal (DCT)')
plt.title('Original and Reconstructed Signal (DCT)')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)

plt.subplot(3, 1, 3)
plt.stem(n, s2, linefmt='b-', markerfmt='bo', basefmt=' ', label='Original signal')
plt.stem(n, s2_dst_reconstructed, linefmt='r--', markerfmt='ro', basefmt=' ', label='Reconstructed signal (DST)')
plt.title('Original and Reconstructed Signal (DST)')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
